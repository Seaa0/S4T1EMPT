```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_csv("datasets/preprocessedDataset.csv")
sns.pairplot(dataset, y_vars=['sleep duration','quality of sleep'], x_vars = [col for col in dataset.columns if col not in ['sleep duration', 'quality of sleep']])
```
